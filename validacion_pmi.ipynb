```python
setx = set(string_list1)
set1 = set()
for word_k in setx:
    set1.add((word_k,string_list1.count(word_k)))
list_words1 = pd.DataFrame(set1)
list_words1.columns = ['word','count']
list_words1 = list_words1.sort_values(['count'], ascending = False)
N = len(list_words1)
N
```

```python
#a partir del vocabulario combinamos para generar tuplas
set_temp=set()
for a in vocabulario[:]:
    for b in vocabulario[:]:
        t=(a,b)
        if(t not in set_temp):
            tupla=(b,a)
            if(tupla not in set_temp):#set_temp.add(t)
                set_temp.add(t)
len(set_temp)
```

Fórmula para el calculo
$PMI(w_i , w_j) = log_2 \frac{ C(w_i , w_j) * M }{ C(w_i)*C(w_j) }$

```python
#Ahora vamos a ejecutar el calculo del pmi para cada palabra
import time
import numpy as np
start = time.time()
diccionario_conteo_word_word=pmi.obtener_dicc_conteo(11, df,diccionario_conteo_word_word1)
print("Se obtuvo el conteo de word a word")
diccionario_PMI_word_word=pmi.obtener_dicc_PMI(diccionario_conteo_word_word,diccionario_conteo_words,M)
print("PMI obtenido")
end = time.time()
print(end-start)
```
